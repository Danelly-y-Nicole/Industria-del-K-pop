# **TIPO DE APRENDIZAJE AUTOMÁTICO A UTILIZAR**

---



El problema de predicción del género de los ***ídolos del K-pop*** mediante análisis de datos y técnicas de aprendizaje automático se enmarca dentro de la categoría de **clasificación supervisada** en el ámbito del aprendizaje automático.

La ***clasificación*** supervisada es un tipo de problema en el que el objetivo es asignar a cada instancia de datos una etiqueta o categoría predefinida a partir de ejemplos previamente etiquetados en un conjunto de entrenamiento. En este caso, las etiquetas son los géneros "masculino" y "femenino", y el objetivo es entrenar un ***modelo de aprendizaje*** automático utilizando datos de ídolos con géneros conocidos para luego predecir el género de nuevos ídolos.

El enfoque de ***Regresión Logística*** utilizado en este análisis es una técnica específica dentro de la clasificación supervisada que se emplea para problemas de clasificación binaria, donde se buscan dos categorías exclusivas. La Regresión Logística aprende a establecer una relación entre las variables predictoras (como la fecha de nacimiento, el lugar de origen, la agencia y la fecha de debut) y la variable objetivo (el género) para hacer predicciones precisas.

***En resumen***, el problema de predicción del género de los ídolos del K-pop se encuadra en la clasificación supervisada, un tipo común de problema que se resuelve mediante técnicas de aprendizaje automático, y se aborda específicamente a través de la aplicación de Regresión Logística para realizar predicciones de género basadas en características de los ídolos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
from datetime import datetime
```

```python
data = pd.read_csv('kpopidolsv3.csv')
```

### Data Cleansing and Data Prep

```python
#check missing ratio
data_na = (data.isnull().sum() / len(data)) * 100
data_na = data_na.drop(data_na[data_na == 0].index).sort_values(ascending=False)[:30]
missing_data = pd.DataFrame({'Missing Ratio' :data_na})
missing_data.head(20)
```

```python
# Handle duplicates
duplicate_rows_data = data[data.duplicated()]
print("number of duplicate rows: ", duplicate_rows_data.shape)
```

```python
# Loop through each column and count the number of distinct values
for column in data.columns:
    num_distinct_values = len(data[column].unique())
    print(f"{column}: {num_distinct_values} distinct values")
```

```python
data['Debut'] = data['Debut'].replace('0/01/1900', pd.NA)
```

```python
data['Date of Birth'] = pd.to_datetime(data['Date of Birth'], format='%d/%m/%Y')
data['Debut'] = pd.to_datetime(data['Debut'], format='%d/%m/%Y')

data['age'] = (datetime.now() - data['Date of Birth']).astype('<m8[Y]')
data['Debut Age'] = (data['Debut'] - data['Date of Birth']).astype('<m8[Y]')

data.head()
```

#### Estatura

```python
# Separate data for males and females
male_data = data[data['Gender'] == 'M']
female_data = data[data['Gender'] == 'F']

# Set color palette
sns.set_palette('pastel')

# Plotting distribution with average for males and females
sns.histplot(data=male_data, x='Height', kde=True, color='skyblue', label='Male')
sns.histplot(data=female_data, x='Height', kde=True, color='lightpink', label='Female')
plt.axvline(x=male_data['Height'].mean(), color='blue', linestyle='dashed', linewidth=1.5, label='Male Avg')
plt.axvline(x=female_data['Height'].mean(), color='red', linestyle='dashed', linewidth=1.5, label='Female Avg')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height Distribution - Males vs. Females')
plt.legend()
plt.show()
```
